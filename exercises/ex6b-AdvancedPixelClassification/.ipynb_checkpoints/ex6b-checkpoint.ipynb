**Exercise 3**

For each binary training ROI find the corresponding training examples in ImgT1 and ImgT2. Later these will be extracted for LDA training.

Tips: If you are a MATLAB-like programming lover, you may use the np.argwhere() function appropriately to return the index to voxels in the image full filling e.g. intensity values >0 hence belong to a given class. Name the index variables qC1 and qC2, respectively.

```python
# Initialize empty lists to store training examples
training_examples_C1_T1 = []
training_examples_C1_T2 = []
training_examples_C2_T1 = []
training_examples_C2_T2 = []

# Iterate through each pixel in the binary masks
for i in range(C1.shape[0]):
    for j in range(C1.shape[1]):
        if C1[i, j] == 1:  # If the pixel belongs to the WM ROI
            training_examples_C1_T1.append(ImgT1[i, j])
            training_examples_C1_T2.append(ImgT2[i, j])
        if C2[i, j] == 1:  # If the pixel belongs to the GM ROI
            training_examples_C2_T1.append(ImgT1[i, j])
            training_examples_C2_T2.append(ImgT2[i, j])

# Convert lists to numpy arrays for further processing
training_examples_C1_T1 = np.array(training_examples_C1_T1)
training_examples_C1_T2 = np.array(training_examples_C1_T2)
training_examples_C2_T1 = np.array(training_examples_C2_T1)
training_examples_C2_T2 = np.array(training_examples_C2_T2)

# Now you have training examples from ImgT1 and ImgT2 corresponding to C1 (WM) and C2 (GM) ROIs.
```

```python
plt.figure(figsize=(12, 8))

# Histogram for training_examples_C1_T1
plt.subplot(2, 2, 1)
plt.hist(training_examples_C1_T1, bins=50, color='blue', alpha=0.7)
plt.title('Histogram for C1 in T1')
plt.xlabel('Pixel Values')
plt.ylabel('Frequency')

# Histogram for training_examples_C1_T2
plt.subplot(2, 2, 2)
plt.hist(training_examples_C1_T2, bins=50, color='green', alpha=0.7)
plt.title('Histogram for C1 in T2')
plt.xlabel('Pixel Values')
plt.ylabel('Frequency')

# Histogram for training_examples_C2_T1
plt.subplot(2, 2, 3)
plt.hist(training_examples_C2_T1, bins=50, color='red', alpha=0.7)
plt.title('Histogram for C2 in T1')
plt.xlabel('Pixel Values')
plt.ylabel('Frequency')

# Histogram for training_examples_C2_T2
plt.subplot(2, 2, 4)
plt.hist(training_examples_C2_T2, bins=50, color='purple', alpha=0.7)
plt.title('Histogram for C2 in T2')
plt.xlabel('Pixel Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Exercise 4**

Make a training data vector (X) and target class vector (T) as input for the LDA() function. T and X should have the same length of data points.

X: Training data vector should first include all data points for class 1 and then the data points for class 2. Data points are the two input features ImgT1, ImgT2

T: Target class identifier for X where '0' are Class 1 and a '1' is Class 2.

Tip: Read the documentation of the provided LDA function to understand the expected input dimensions.

```python
# Combine the data points for class 1 and class 2
class_1_data = np.column_stack((training_examples_C1_T1, training_examples_C1_T2))
class_2_data = np.column_stack((training_examples_C2_T1, training_examples_C2_T2))

# Create the training data vector X
X = np.vstack((class_1_data, class_2_data))

# Create the target class vector T
# '0' represents class 1, '1' represents class 2
T_class_1 = np.zeros(len(training_examples_C1_T1))
T_class_2 = np.ones(len(training_examples_C2_T1))
T = np.concatenate((T_class_1, T_class_2))

# X and T are now ready to be used as input for the LDA function.
```

**Exercise 6**

Train the LDA

```python
def LDA(X, y):
    """
    Linear Discriminant Analysis.

    A classifier with a linear decision boundary, generated by fitting class 
    conditional densities to the data and using Bayes’ rule.
    Assumes equal priors among classes

    Parameters
    ----------
    X : array-like of shape (n_samples, n_features)
        Training data
    y : array-like of shape (n_samples,)
        Target values.

    Returns
    -------
    W : array-like of shape (n_classes, n_features+1)
        Weights for making the projection. First column is the constants.

    Last modified: 11/11/22, mcbo@dtu.dk
    """
    
    # Determine size of input data
    n, m = X.shape
    # Discover and count unique class labels
    class_label = np.unique(y)
    k = len(class_label)

    # Initialize
    n_group     = np.zeros((k,1))     # Group counts
    group_mean  = np.zeros((k,m))     # Group sample means
    pooled_cov  = np.zeros((m,m))     # Pooled covariance
    W          = np.zeros((k,m+1))    # Model coefficients

    for i in range(k):
        # Establish location and size of each class
        group      = np.squeeze(y == class_label[i])
        n_group[i]  = np.sum(group.astype(np.double))
        
        # Calculate group mean vectors
        group_mean[i,:] = np.mean(X[group,:], axis = 0)
        
        # Accumulate pooled covariance information
        pooled_cov = pooled_cov + ((n_group[i] - 1) / (n - k) ) * np.cov(X[group,:], rowvar = False)
    
    # Assign prior probabilities
    prior_prob = n_group / n

    # Loop over classes to calculate linear discriminant coefficients
    for i in range(k):
        # Intermediate calculation for efficiency
        temp = group_mean[i,:][np.newaxis] @ np.linalg.inv(pooled_cov)
        
        # Constant
        W[i,0] = -0.5 * temp @ group_mean[i,:].T + np.log(prior_prob[i])
        
        # Linear
        W[i,1:] = temp

    return W
```

```python
W = LDA(X, T)
```

```python
Xall = np.c_[ImgT1.ravel(), ImgT2.ravel()]
Y = np.c_[np.ones((len(Xall), 1)), Xall] @ W.T
```

**Exercise 8**

Perform multi-modal classification: Calculate the posterior probability i.e. 
of a data point belonging to class 1

Note: Using Bayes [Eq 1]: Since 
 is the log of the posterior probability [Eq2] we take 
 to get 
 and divide with the marginal probability 
 as normalisation factor.

```python
PosteriorProb = np.clip((np.exp(Y) / np.sum(np.exp(Y),1)[:, np.newaxis]), 0, 1)
```

**Exercise 9**

Apply segmentation: Find all voxles in the T1w and T2w image with 
 as belonging to Class 1. You may use the np.where() function. Similarly, find all voxels belonging to class 2.

```python
# Supongamos que PosteriorProb es la matriz de probabilidades posteriores
# donde PosteriorProb[i, 0] es la probabilidad de que el píxel i pertenezca a la Clase 1
# y PosteriorProb[i, 1] es la probabilidad de que pertenezca a la Clase 2.

# Definir un umbral para clasificar los píxeles en las clases.
umbral_clase_1 = 0.5  # Ajusta el umbral según tus necesidades.
umbral_clase_2 = 0.5  # Ajusta el umbral según tus necesidades.

# Segmentar las imágenes T1w y T2w en las clases 1 y 2.
clase_1_indices_T1w = np.where(PosteriorProb[:, 0] > umbral_clase_1)
clase_2_indices_T1w = np.where(PosteriorProb[:, 1] > umbral_clase_2)

clase_1_indices_T2w = np.where(PosteriorProb[:, 0] > umbral_clase_1)
clase_2_indices_T2w = np.where(PosteriorProb[:, 1] > umbral_clase_2)

# Ahora, clase_1_indices_T1w y clase_2_indices_T1w contienen los índices de los píxeles
# que pertenecen a las clases 1 y 2 en la imagen T1w, respectivamente.

# Puedes utilizar estos índices para acceder a los píxeles correspondientes en las imágenes originales.
```
