Six photos have been taken. They are named orchid001.png -
orchid006.png. The dimensions of each photo are (W=400, H=533 pixels).
In order to do some statistical analysis of the images, the average image
of the six photos is computed. Finally, a threshold of 150 is applied to the
average image so pixels above the threshold are considered as
foreground and the rest as background. Which is the resulting image?

```python
import glob

def create_u_byte_image_from_vector(im_vec, height, width):
    min_val = im_vec.min()
    max_val = im_vec.max()

    # Transform to [0, 1]
    im_vec = np.subtract(im_vec, min_val)
    im_vec = np.divide(im_vec, max_val - min_val)
    im_vec = im_vec.reshape(height, width)
    im_out = img_as_ubyte(im_vec)
    return im_out

# Import all the image files from the directory
preprocess_dir = r"C:\Users\jdiaz\Desktop\DTU_ImageAnalysis\Exams\Fall2021\Orchid"
files = glob.glob(f"{preprocess_dir}/*.png")

# We create a matrix containing all the images and flatten it
height, width = io.imread(files[0]).shape
data_matrix = np.zeros((len(files), height * width))
for i, f in enumerate(files):
    data_matrix[i, :] = io.imread(f).flatten()
    
# Compute the average image from the ones we have

average_data = np.mean(data_matrix, axis=0)

# Create an actual image from it, because it was flatten
average_image = create_u_byte_image_from_vector(average_data, height, width)
bin_img = average_image > 150
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
plt.imshow(bin_img,cmap="gray")
ax.set_title("Mean Image")
ax.set_axis_off()
plt.show()
```

A principal component analysis (PCA) should be done on these data.
How many principal components are needed to explain at least 95% of
the variation in the data set?

```python
# Exercise 1
in_dir = r"C:\Users\jdiaz\Desktop\DTU_ImageAnalysis\Exams\Fall2021"
txt_name = r"\pizza.txt"

import numpy as np
pizza_data = np.loadtxt(in_dir + txt_name, comments="%")
# x is a matrix with 50 rows and 4 columns

def cov_matrix_calc(x):
    mn = np.mean(x,axis=0)
    data = x-mn
    dot = np.matmul(data.T,data)
    cx = dot/(len(x)-1)
   
    return cx

c_x = cov_matrix_calc(pizza_data)

values, vectors = np.linalg.eig(c_x)
```

```python
var_ratio = []
nums = np.arange(pizza_data.shape[1])
for num in nums:
    pca = decomposition.PCA(n_components=num)
    pca.fit(pizza_data)
    var_ratio.append(np.sum(pca.explained_variance_ratio_))
plt.figure(figsize=(10,5),dpi=150)
plt.grid()
plt.plot(nums,var_ratio,marker='o')
plt.xlabel('n_components')
plt.ylabel('Explained variance ratio')
plt.title('n_components vs. Explained Variance Ratio')
```

What is the DICE score between your segmentation and the ground truth
segmentation?

```python
def DICE_COE(mask1, mask2):
    intersect = np.sum(mask1*mask2)
    fsum = np.sum(mask1)
    ssum = np.sum(mask2)
    dice = (2 * intersect ) / (fsum + ssum)
    dice = np.mean(dice)
    dice = round(dice, 3) # for easy reading
    return dice    

img_path = r"C:\Users\jdiaz\Desktop\DTU_ImageAnalysis\Exams\Fall2021\boxROI.png"
img_roi = io.imread(img_path)

print(img.dtype)
print(img_roi.dtype)
DICE_COE(img/255,img_roi/255)
```

The photo called water_gray.png is filtered using a 3x3 median filter. A
linear histogram stretching is then performed on the filtered image so
the new image has a maximum pixel value of 230 and a minimum pixel
value of 12. What is the pixel value at (row=20, column=20) in the new
image (a 1-based matrix coordinate system is used)?


```python
img_path = r"C:\Users\jdiaz\Desktop\DTU_ImageAnalysis\Exams\Fall2021\water_gray.png"
img = io.imread(img_path)

#img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

filter_size = 3
img_median_filter = median(img, square(filter_size))

def histogram_stretch(img_in):
    """
    Stretches the histogram of an image 
    :param img_in: Input image
    :return: Image, where the histogram is stretched so the min values is 0 and the maximum value is 255
    """
    # Convert the input image to a floating-point representation
    img_float = img_as_float(img_in)
    
    # Find the minimum and maximum values in the image
    min_val = img_float.min()
    max_val = img_float.max()
    
    # Define the desired minimum and maximum values
    min_desired = 12/255
    max_desired = 230/255
    
    # Stretch the histogram
    img_out = (img_float - min_val) / (max_val - min_val)  # Normalize to [0, 1]
    img_out = (img_out * (max_desired - min_desired)) + min_desired  # Stretch to [min_desired, max_desired]
    
    # Convert the stretched image back to unsigned byte format (0-255)
    img_out = img_as_ubyte(img_out)
    
    return img_out

stretched_image = histogram_stretch(img_median_filter)

stretched_image[19,19]
```

```python
# Compute the pixel value ranges in a parametric classifier

# Define the training data for each class (gray values)
category1_values = np.array([68, 65, 67])
category2_values = np.array([70, 80, 75])
category3_values = np.array([77, 92, 89])

# Estimate the parameters (mean and variance) for each class
mean_category1 = np.mean(category1_values)
variance_category1 = np.var(category1_values,ddof=1)
mean_category2 = np.mean(category2_values)
variance_category2 = np.var(category2_values,ddof=1)
mean_category3 = np.mean(category3_values)
variance_category3 = np.var(category3_values,ddof=1)

# Define the number of standard deviations to include in the range
num_std_devs = 1  # You can adjust this value as needed

# Define the Gaussian PDFs for each class
pdf_category1 = stats.norm(loc=mean_category1, scale=np.sqrt(variance_category1))
pdf_category2 = stats.norm(loc=mean_category2, scale=np.sqrt(variance_category2))
pdf_category3 = stats.norm(loc=mean_category3, scale=np.sqrt(variance_category3))

# Define a function to find the intersection points
def find_intersection(x):
    return pdf_category1.pdf(x) - pdf_category2.pdf(x)

# Use fsolve to find the pixel intensity value where Grass and Road PDFs intersect
intersection_point_categories12 = fsolve(find_intersection, x0=(mean_category1 + mean_category2) / 2)

# Define a function to find the intersection points
def find_intersection(x):
    return pdf_category2.pdf(x) - pdf_category3.pdf(x)

# Use fsolve to find the pixel intensity value where Grass and Road PDFs intersect
intersection_point_categories23 = fsolve(find_intersection, x0=(mean_category2 + mean_category3) / 2)


category1_range = (0,round(intersection_point_categories12[0]))
category2_range = (round(intersection_point_categories12[0]),round(intersection_point_categories23[0]))
category3_range = (round(intersection_point_categories23[0]),255)


# Display the extreme values of the class ranges
print("Class Ranges:")
print(f"category1 Range: {category1_range}")
print(f"category2 Range: {category2_range}")
print(f"category3 Range: {category3_range}")
```
