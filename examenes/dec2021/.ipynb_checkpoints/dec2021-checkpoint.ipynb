# el ejercicio de donald

Two similar images have been acquired (donald_1.png and
donald_2.png). Six corresponding landmarks have been placed on the
two images and are stored in the files donaldfixedPoints.mat and
donaldmovingPoints.mat.
A similarity transform (translation, rotation, and scaling) has been
performed that aligns the moving points to the fixed points. The
computed transform is applied to the donald_2.png photo. What is the
RGB value at position (row=300, column=300, in a 1-based matrix
coordinate system) in the transformed image?

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_ubyte
from skimage.transform import SimilarityTransform, warp,matrix_transform

donald_1_path = r"./data/Donald/donald_1.png"
donald_2_path = r"./data/Donald/donald_2.png"

src_img = io.imread(donald_2_path)
print(src_img.dtype)
dst_img = io.imread(donald_1_path)

from scipy.io import loadmat

moving_path = r"./data/Donald/donaldmovingPoints.mat"
fixed_path = r"./data/Donald/donaldfixedPoints.mat"

# Load the .mat file
moving_pts = loadmat(moving_path)

# Access variables from the loaded data
variable_name = 'movingPoints'
moving_pts = moving_pts[variable_name]

# Load the .mat file
fixed_pts = loadmat(fixed_path)
print(fixed_pts)
# Access variables from the loaded data
variable_name = 'fixedPoints'
fixed_pts = fixed_pts[variable_name]

# source
#moving_pts = np.array([[208.297427652733, 66.8183279742764], [129.519292604502, 131.127009646302], [45.1141479099678, 206.689710610932], [265.371382636656, 455.885852090032], [352.188102893891, 382.734726688103], [441.416398713826, 307.975884244373]])
# Destination
#fixed_pts = np.array([[363.442122186495, 58.7797427652732], [235.628617363344, 60.3874598070739], [112.638263665595, 63.6028938906752], [101.384244372990, 447.043408360129], [225.982315112540, 450.258842443730], [368.265273311897, 454.278135048232]])

tform = SimilarityTransform()
tform.estimate(moving_pts, fixed_pts)
src_transform = matrix_transform(moving_pts, tform.params)

# The found transform can be applied to the source image
warped = warp(src_img, tform.inverse)

def show_comparison(original, transformed, transformed_name):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True, sharey=True)
    ax1.imshow(original)
    ax1.set_title('Original')
    ax1.axis('off')
    ax2.imshow(transformed)
    ax2.set_title(transformed_name)
    ax2.axis('off')
    plt.show()

show_comparison(src_img, warped, 'Landmark based transformation')

# Get the RGB value at position (row=300, column=300) in the transformed image
# Assuming img contains pixel values in the range [0, 1]
rgb_value = warped[299, 299]
print(rgb_value)
print(warped.dtype)

# Convert the transformed image to uint8 format
donald2t_uint8 = img_as_ubyte(warped)

# Extract the RGB values at the specified pixel location
row = 299  # 0-based coordinate
column = 299  # 0-based coordinate

resR = donald2t_uint8[row, column, 0]
resG = donald2t_uint8[row, column, 1]
resB = donald2t_uint8[row, column, 2]

print(f"Red: {resR}, Green: {resG}, Blue: {resB}")
```

```python
def DICE_COE(mask1, mask2):
    intersect = np.sum(mask1*mask2)
    fsum = np.sum(mask1)
    ssum = np.sum(mask2)
    dice = (2 * intersect ) / (fsum + ssum)
    dice = np.mean(dice)
    dice = round(dice, 3) # for easy reading
    return dice    

dice_result = DICE_COE(img/255,img_roi/255)
print(dice_result)
```

```python
img_path = r"./data/water_gray.png"
img = io.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)


filter_size = 3
img_median_filter = median(img, square(filter_size))
```

```python
def histogram_stretch(img_in):
    # ojo aqui al formato de las imagenes de entrada
    # si me dan el max o min desired en px tengo que tener la imagen en uint8
    # si me dan el max o min desired en float tengo que tener la img en float
    # quitar o poner el 255 de abajo si necesario      
    """
    Stretches the histogram of an image 
    :param img_in: Input image
    :return: Image, where the histogram is stretched so the min values is 0 and the maximum value 255
    """
    # img_as_float will divide all pixel values with 255.0
    img_float = img_as_float(img_in)
    min_val = img_float.min()
    max_val = img_float.max()
        
    
    min_desired = 12/255
    max_desired = 230/255
	
    img_out = ( (max_desired-min_desired) / (max_val-min_val)) * ((img_float-min_val) + min_desired)
    
    # img_as_ubyte will multiply all pixel values with 255.0 before converting to unsigned byte
    plt.figure()
    
    plt.subplot(221)
    plt.hist(img_float.ravel())
    plt.title("Histogram before stretching")
    
    plt.subplot(222)
    io.imshow(img_float)
    plt.title("Image before stretching")
    
    plt.subplot(223)
    plt.hist(img_out.ravel())
    plt.title("Histogram after stretching")
    
    plt.subplot(224)
    io.imshow(img_out)
    plt.title("Image after stretching")
    
    plt.show()
    print(f"Max before stretching: {img_float.max():.2f}")
    print(f"Min before stretching: {img_float.min():.2f}\n")
    
    print("Max after stretching:",img_out.max())
    print("Min after stretching:",img_out.min())
    print()
    print(f"Average pixel value after stretching: {img_out.mean():.3f}\n" )
    return img_as_ubyte(img_out)
```

# classify grass, road and sky

```python
import numpy as np
from scipy import stats, optimize

def compute_pixel_value_ranges(category1_values, category2_values, category3_values):
    # Estimate the parameters (mean and variance) for each class
    mean_category1 = np.mean(category1_values)
    variance_category1 = np.var(category1_values, ddof=1)
    mean_category2 = np.mean(category2_values)
    variance_category2 = np.var(category2_values, ddof=1)
    mean_category3 = np.mean(category3_values)
    variance_category3 = np.var(category3_values, ddof=1)

    # Define the number of standard deviations to include in the range
    num_std_devs = 1  # You can adjust this value as needed

    # Define the Gaussian PDFs for each class
    pdf_category1 = stats.norm(loc=mean_category1, scale=np.sqrt(variance_category1))
    pdf_category2 = stats.norm(loc=mean_category2, scale=np.sqrt(variance_category2))
    pdf_category3 = stats.norm(loc=mean_category3, scale=np.sqrt(variance_category3))

    # Define a function to find the intersection points
    def find_intersection12(x):
        return pdf_category1.pdf(x) - pdf_category2.pdf(x)

    def find_intersection23(x):
        return pdf_category2.pdf(x) - pdf_category3.pdf(x)

    # Use fsolve to find the pixel intensity values where PDFs intersect
    intersection_point_categories12 = optimize.fsolve(find_intersection12, x0=(mean_category1 + mean_category2) / 2)
    intersection_point_categories23 = optimize.fsolve(find_intersection23, x0=(mean_category2 + mean_category3) / 2)

    # Define the pixel intensity ranges for each category
    category1_range = (0, round(intersection_point_categories12[0]))
    category2_range = (round(intersection_point_categories12[0]), round(intersection_point_categories23[0]))
    category3_range = (round(intersection_point_categories23[0]), 255)

    # Return the pixel intensity ranges
    return category1_range, category2_range, category3_range
```

```python
# Example usage:
category1_values = np.array([68, 65, 67])
category2_values = np.array([70, 80, 75])
category3_values = np.array([77, 92, 89])

result = compute_pixel_value_ranges(category1_values, category2_values, category3_values)

# Display the extreme values of the class ranges
print("Class Ranges:")
print(f"category1 Range: {result[0]}")
print(f"category2 Range: {result[1]}")
print(f"category3 Range: {result[2]}")
```
